```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set style
sns.set(style="whitegrid")
```

```python
import os

# Create directory if it doesn't exist
os.makedirs("visualizations", exist_ok=True)
```

```python
item_units = df.groupby("item_sold")["units_sold"].sum().sort_values()

plt.figure(figsize=(10, 6))
sns.barplot(x=item_units.values, y=item_units.index, hue=item_units.index, palette="Blues_d", legend=False)
plt.title("Total Units Sold by Item")
plt.xlabel("Units Sold")
plt.ylabel("Item Sold")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="gross_profit", y="company_name", data=df.sort_values("gross_profit"), hue="company_name", palette="Greens_d", legend=False)
plt.title("Gross Profit by Company")
plt.xlabel("Gross Profit (USD)")
plt.ylabel("Company Name")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["cost_per_unit_usd"], kde=True, color="orange", bins=10)
plt.title("Distribution of Cost per Unit")
plt.xlabel("Cost per Unit (USD)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="bill_amount", y="gross_profit", hue="company_name", palette="tab10")
plt.title("Bill Amount vs Gross Profit")
plt.xlabel("Bill Amount (USD)")
plt.ylabel("Gross Profit (USD)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.select_dtypes(include=["number"]).corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
item_units = df.groupby("item_sold")["units_sold"].sum().sort_values()

plt.figure(figsize=(10, 6))
sns.barplot(x=item_units.values, y=item_units.index, hue=item_units.index, palette="Blues_d", legend=False)
plt.title("Total Units Sold by Item")
plt.xlabel("Units Sold")
plt.ylabel("Item Sold")
plt.tight_layout()
plt.show()
```
